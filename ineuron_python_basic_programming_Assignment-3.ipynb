```python
# 1. Write a Python Program to Check if a Number is Positive, Negative or Zero?


x = int(input("enter the value: "))
if x > 0:
    print("It is a positive value. ")
elif x < 0:
    print("It is a negative value.")
else:
    print("This a Zero value")
```

```python
# 2. Write a Python Program to Check if a Number is Odd or Even?

x = int(input("enter the value:"))

if x%2==0:
    print("it is a even number: {}".format(x))
else:
    print("it is a odd number: {}".format(x))
```

```python
# 3. Write a Python Program to Check Leap Year?

Leap_year = int(input("Enter a year: "))

if Leap_year%4==0:
    print("{} year is a Leap year".format(Leap_year))
else:
    print("{} is not a Leap Year".format(Leap_year))
```

```python
# 4. Write a Python Program to Check Prime Number?


prime_number = int(input("Enter a number: "))

for i in range(2,prime_number):
    if i%2 == 0:
        print("{} is a prime number: ".format(prime_number))
        break
else: 
    print("{} is not a prime number: ".format(prime_number))
```

```python
# 5. Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?


x = int(input("Enter a number 1: "))
y = int(input("Enter a number 2: "))
for num in range(x,y):
    if num>1:
        for i in range(2,num):
            if (num%i)==0:
                print(num, "is not a prime")
                break
            else:
                print(num, "is a prime number")
```
